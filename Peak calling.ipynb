
# <center>Peak detection with simple thresholding</center>

## ===========================================================

## Peak detection channel 1

## ===========================================================

```python
from scipy.signal import find_peaks

th1 = np.mean(y1)

gp_ch1 = []

peaks1, _ = find_peaks(y1, height=th1)

xp1 = []
yp1 = []
yp2_1 = []

for p in peaks1:
    xp1.append(x1[p])
    yp1.append(y1[p])
    yp2_1.append(y2[p])
    
    gp_ch1.append((y1[p] - y2[p])/(y1[p] + y2[p]))
    

plt.plot(x1,y1)
plt.plot(xp1,yp1, 'x')
plt.show()

plt.hist(gp_ch1, bins = int(np.sqrt(len(y1))))
plt.show()
```

## ===========================================================

## Peak detection channel 2

## ===========================================================

```python
th2 = np.mean(y2)

gp_ch2 = []

peaks1, _ = find_peaks(y2, height=th2)

xp2 = []
yp2 = []
yp1_2 = []

for p in peaks1:
    xp2.append(x2[p])
    yp2.append(y2[p])
    yp1_2.append(y1[p])
    
    gp_ch2.append((y1[p] - y2[p])/(y1[p] + y2[p]))

plt.plot(x2,y2)
plt.plot(xp2,yp2, 'x')
plt.show()

plt.hist(gp_ch2, bins = int(np.sqrt(len(y1))))
plt.show()
```

## ===========================================================

## GP calculation

## ===========================================================

```python
print("Detected in different channels")



plt.hist(gp_ch1, bins = int(np.sqrt(len(y1))), label = 'channel 1')
plt.hist(gp_ch2, bins = int(np.sqrt(len(y1))), label = 'channel 2')
plt.legend(loc='upper left')
plt.show()

gp_all = gp_ch1 + gp_ch2

print("Detected in different channels cumulative")

plt.hist(gp_all, bins = int(np.sqrt(len(y1))))

plt.show()

```
