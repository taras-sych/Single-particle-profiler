# Normalization:
    - mean, 
    - z scale 

# =============================================

```python
from scipy import stats

normalization_index = "z-score"
#normalization_index = "mean"

print("Positive:")
plt.plot (x1, y1)
plt.plot (x2, y2)
plt.show()

print("Negative:")
plt.plot (x3, y4)
plt.plot (x4, y4)
plt.show()

if normalization_index == "z-score":

    y1z = stats.zscore(y1)
    y2z = stats.zscore(y2)

    y1 = y1z - min(y1z)
    y2 = y2z - min(y2z)

if normalization_index == "mean":

    y1 = y1/np.mean(y1)
    y2 = y2/np.mean(y2)
    
    
if normalization_index == "z-score":

    y3z = stats.zscore(y3)
    y4z = stats.zscore(y4)

    y3 = y3z - min(y3z)
    y4 = y4z - min(y4z)

if normalization_index == "mean":

    y3 = y3/np.mean(y3)
    y4 = y4/np.mean(y4)

    

    
print("Positive:")
plt.plot (x1, y1)
plt.plot (x2, y2)
plt.show()

print("Negative:")
plt.plot (x3, y4)
plt.plot (x4, y4)
plt.show()

"""
plt.scatter(y1,y2,color='red',label = 'positive')
plt.scatter(y3,y4,color='blue', label = 'negative')
plt.xlabel('channel 1')
plt.ylabel('channel 2')
plt.legend(loc='upper right')
plt.show()

plt.yscale ('symlog')
plt.xscale ('symlog')
plt.scatter(y1,y2,color='red',label = 'positive')
plt.scatter(y3,y4,color='blue', label = 'negative')

plt.xlabel('channel 1')
plt.ylabel('channel 2')
#plt.ylim(0.1, 10)
#plt.xlim(0.1, 10)
plt.legend(loc='upper left')
plt.show()
"""
```

```python
from scipy.signal import find_peaks

th1 = np.mean(y1)
th2 = np.mean(y2)
th3 = np.mean(y3)
th4 = np.mean(y4)

"""th1 = 7
th2 = 7
th3 = 5
th4 = 5"""

th1 = 0
th2 = 0
th3 = 0
th4 = 0

peaks1, _ = find_peaks(y1, height=th1)

xp1 = []
yp1 = []
yp2_1 = []
for p in peaks1:
    xp1.append(x1[p])
    yp1.append(y1[p])
    yp2_1.append(y2[p])


peaks2, _ = find_peaks(y2, height=th2)

xp2 = []
yp2 = []
yp1_2 = []

for p in peaks2:
    xp2.append(x2[p])
    yp2.append(y2[p])
    yp1_2.append(y1[p])
    

    
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(xp1, yp1, 'x')
plt.plot(xp2, yp2, 'x')
plt.show()
    
peaks3, _ = find_peaks(y3, height=th3)

xp3 = []
yp3 = []
yp4_3 = []
for p in peaks3:
    xp3.append(x3[p])
    yp3.append(y3[p])
    yp4_3.append(y4[p])


peaks4, _ = find_peaks(y4, height=th4)

xp4 = []
yp4 = []
yp3_4 = []

for p in peaks4:
    xp4.append(x4[p])
    yp4.append(y4[p])
    yp3_4.append(y3[p])
    
    
plt.plot(x3,y3)
plt.plot(x4,y4)
plt.plot(xp3, yp3, 'x')
plt.plot(xp4, yp4, 'x')
plt.show()


```
